````python
import numpy as np

# Entrada para la neurona
X = np.array ([
    [0,0],
    [0,1],
    [1,0],
    [1,1],
])

# Salidas esperadas para OR
Y = np.array([0,1,1,1])

# Pesos iniciales aleatorios para el Perceptron
W = np.random.randn(2)

# Tasa de aprendizaje para el entrenamiento
learning_rate = 0.1

# N√∫mero de √©pocas de entrenamiento
epochs = 100

class Perceptron:
    def __init__(self, weights, learning_rate):
        """
        Constructor del Perceptron.
        weights -> vector pesos iniciales (aleatorios)
        learning_rate -> tasa de aprendizaje para ajustar pesos
        """
        self.weights = weights
        self.learning_rate = learning_rate
        # En Perceptron, el umbral se maneja como un peso m√°s (bias)
        self.bias = np.random.randn()

    def predict(self, X):
        """Calcula la salida del Perceptron para cada fila de entradas X."""
        predictions = []
        for i in range(X.shape[0]):
            # Calcular la suma ponderada: w1*x1 + w2*x2 + bias
            weighted_sum = np.dot(self.weights, X[i]) + self.bias
            print(f"Input: {X[i]}, Weighted Sum: {weighted_sum:.4f}")
            # Funci√≥n de activaci√≥n escal√≥n (como MCP pero con bias)
            if weighted_sum >= 0:
                predictions.append(1)
            else:
                predictions.append(0)
        return predictions

    def train(self, X, Y, epochs):
        """Entrena el Perceptron ajustando pesos y bias."""
        print("Iniciando entrenamiento...")
        print(f"Pesos iniciales: {self.weights}, Bias inicial: {self.bias:.4f}")

        for epoch in range(epochs):
            total_error = 0
            for i in range(X.shape[0]):
                # Calcular predicci√≥n
                weighted_sum = np.dot(self.weights, X[i]) + self.bias
                prediction = 1 if weighted_sum >= 0 else 0

                # Calcular error
                error = Y[i] - prediction
                total_error += abs(error)

                # Actualizar pesos y bias (Regla de aprendizaje del Perceptron)
                self.weights += self.learning_rate * error * X[i]
                self.bias += self.learning_rate * error

            # Mostrar progreso cada 10 √©pocas
            if epoch % 10 == 0:
                print(f"√âpoca {epoch}: Error total = {total_error}")

# Instanciamos la clase Perceptron
perceptron = Perceptron(W, learning_rate)

# Entrenamos el Perceptron
perceptron.train(X, Y, epochs)

# Calculamos las predicciones finales
predictions = perceptron.predict(X)

# Mostramos resultado final
print("Predicciones finales:", predictions)
print("Salidas esperadas:   ", Y.tolist())
print("Pesos finales:", perceptron.weights)
print("Bias final:", perceptron.bias)
````

## üîß Partes del C√≥digo

### 1. **Inicializaci√≥n**
```python
W = np.random.randn(2)           # Pesos aleatorios
learning_rate = 0.1              # Velocidad de aprendizaje
```
- **Pesos aleatorios**: Empieza "sin saber nada"
- **Learning rate**: Qu√© tan r√°pido aprende

### 2. **Predicci√≥n**
```python
weighted_sum = w1*x1 + w2*x2 + bias
if weighted_sum >= 0:
    return 1
else:
    return 0
```
- **Suma ponderada**: Entradas √ó Pesos + Bias
- **Activaci√≥n**: Si suma ‚â• 0 ‚Üí 1, si no ‚Üí 0

### 3. **Entrenamiento (La Magia üé©)**
```python
error = respuesta_correcta - predicci√≥n
W += learning_rate * error * entrada
bias += learning_rate * error
```
- **Error**: Diferencia entre lo que dijo y lo que deb√≠a decir
- **Ajuste**: Cambia pesos seg√∫n el error
